```python
import pandas as pd
import numpy as np

trainDF = pd.read_csv("Kaggle Data/test_supplement.csv")
trainDF.head()
```

```python
trainDF['click_time'] = pd.to_datetime(trainDF['click_time'])

trainDF['Hour'] = trainDF['click_time'].dt.hour
trainDF['day_of_week'] = trainDF['click_time'].dt.dayofweek
hr = []

trainDF = trainDF.drop("click_time",axis=1)
trainDF = trainDF.drop("click_id",axis=1)
#trainDF = trainDF.drop("attributed_time",axis=1)

trainDF.head()
```

```python
#trainDF.loc[trainDF.Hour <24 , 'TimeOfDay'] = 'Night'
#trainDF.loc[trainDF.Hour <20 , 'TimeOfDay'] = 'Afternoon'
#trainDF.loc[trainDF.Hour <16 , 'TimeOfDay'] = 'Noon'
#trainDF.loc[trainDF.Hour <12 , 'TimeOfDay'] = 'Morning'
#trainDF.loc[trainDF.Hour <8, 'TimeOfDay'] = 'Dusk'
#trainDF.loc[trainDF.Hour <4, 'TimeOfDay'] = 'Midnight'

trainDF.loc[trainDF.Hour <24 , 'TimeOfDay_N'] = 6
trainDF.loc[trainDF.Hour <20 , 'TimeOfDay_N'] = 5
trainDF.loc[trainDF.Hour <16 , 'TimeOfDay_N'] = 4
trainDF.loc[trainDF.Hour <12 , 'TimeOfDay_N'] = 3
trainDF.loc[trainDF.Hour <8, 'TimeOfDay_N'] = 2
trainDF.loc[trainDF.Hour <4, 'TimeOfDay_N'] = 1
```
